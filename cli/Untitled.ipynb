```unity-kernel
import module namespace p = "http://dgms.io/unity/modules/visualization/plotting";

import module namespace m = "http://www.w3.org/2005/xpath-functions/math";

p:plot2d({
columns:1,
rows:2,
    plots: [
        {
            data:
            [
                {
                    style: "o",
                    mapper: function($x) {$x}
                },
                {
                    style: "-",
                    mapper: function($x) {2*$x}
                }
            ]
        },
        {
            data:
            [
                {
                    style: "o",
                    mapper: function($x) {$x}
                },
                                {
                    style: "o",
                    mapper: function($x) {$x*5}
                },
                                {
                    style: "o",
                    mapper: function($x) {$x*5*5}
                },
                                {
                    style: "o",
                    mapper: function($x) {$x*5*5*5}
                },
                                {
                    style: "o",
                    mapper: function($x) {$x*5*5*5}
                },
                {
                    style: "-",
                    mapper: function($x) {m:sin(2*$x)*5}
                }
            ]
        }
    ]
})
```
